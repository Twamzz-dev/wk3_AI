```python
import bz2
from pathlib import Path

# Use a raw string with Path to avoid unicode-escape errors from backslashes
data = Path(r"C:\Users\USER\.cache\kagglehub\datasets\bittlingmayer\amazonreviews\versions\7\test.ft.txt.bz2")

if not data.exists():
    raise FileNotFoundError(f"Data file not found: {data}")

with bz2.open(data, "rt", encoding="utf-8") as f:
    lines = []
    for _ in range(200):
        try:
            lines.append(next(f))
        except StopIteration:
            break
```

```python
# strip labels and keep text only
texts = [line.split(" ", 1)[1].strip() for line in lines]

# Import necessary libraries
import spacy

# load spacy model
nlp = spacy.load("en_core_web_sm")

# run spacy NER on each review
for text in texts[:5]:
    doc = nlp(text)
    print([(ent.text, ent.label_) for ent in doc.ents])
```

```python
# peek into the reviews and  print them nicely
for i, t in enumerate(texts, 1):
    print(f"Review {i}: {t}\n")
```

```python
# rule-based sentiment analysis

positive_words = ["good", "great", "best", "love", "excellent", "cool"]
negative_words = ["bad", "horrible", "disappointed", "disturbing", "waste", "problem"]

def rule_based_sentiment(text):
    text_lower = text.lower()
    
    pos = sum(word in text_lower for word in positive_words)
    neg = sum(word in text_lower for word in negative_words)

    if pos > neg:
        return "positive"
    elif neg > pos:
        return "negative"
    else:
        return "neutral"

# test the rule-based sentiment analysis
for t in texts[:10]:
    print(t[:60], "->", rule_based_sentiment(t))
```
